**sotools example notebook**

```python
%matplotlib inline
from sotools import enmap, utils
from matplotlib import pyplot as plt
import numpy as np
```

```python
# load maps
emap = enmap.read_map("testmap.fits")
plt.matshow(emap); plt.show()

# write maps
emap.write("testmap_out.fits")
```

```python
# extracting basic map information
print "extract basic map information"
print "area (in steradian) : %f " %emap.area()  
print "mean, std: %f %f" %(emap.mean(), emap.std())
print "map geometry: %s" %str(emap.geometry)
```

```python
# selecting a sub section of a map
delta = 1. * utils.degree

emap_sub = emap.submap([[-delta,-delta],[delta,delta]]) # [[yfrom, xfrom], [yto,xto]]
plt.matshow(emap_sub); plt.show()
```

```python
# Fourier transform 
ftmap  = enmap.fft(emap)     # forward fourier transform
iftmap = enmap.ifft(ftmap)   # inverse fourier transform
```
